```python
import pandas as pd
```

```python
trd=pd.read_csv("student_trainData.csv")
trt=pd.read_csv("student_trainTarget.csv")
td=pd.read_csv("student_testData.csv")
tt=pd.read_csv("student_testTarget.csv")
```

```python
trd.head()
```

```python
trt.head()
```

```python
td.head()
```

```python
tt.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(max_depth=9,criterion ='entropy')
#clf = clf.fit(x_train, y_train)
clf = clf.fit(trd, trt)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(td)
predictions
```

```python
clf.predict_proba(td)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(tt,predictions )
```

```python

from sklearn.metrics import confusion_matrix
confusion_matrix(tt, predictions,labels=[0,1])
```

```python
from sklearn.metrics import precision_score
precision_score(tt, predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(tt, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(tt,predictions))
```
